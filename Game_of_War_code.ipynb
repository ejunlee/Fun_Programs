```python
import random
```

```python
# Card Class
class Card():
# Creating the class of a card in order to build a deck
    def __init__ (self,suit,rank):
        self.suit = suit
        self.rank = rank
        #self.value is an attribute to compare cards with int
        self.value = values[rank]
    '''
    Want to create a string where if you print a card
    it will return the value and suit
    '''
    def __str__ (self):
        return self.rank + " of " + self.suit
#Suit,Rank,Value
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}

#Deck
class Deck():
    
    def __init__(self):
        #creating a list of cards
        self.all_cards = []
        
        for suit in suits:
            for rank in ranks:
                # create the card object
                created_card = Card(suit,rank)
                self.all_cards.append(created_card)

    def shuffle(self):
        
        random.shuffle(self.all_cards)
    
    def deal_one(self):
        return self.all_cards.pop()
#Player Class
'''
players class is used to hold a player's current list of cards
'''
#if you use append on two list then you create a nested list
#instead use the extend method ".extend()"
class Player:
    
    def __init__(self,name):
        self.name = name
        self.all_cards = []
        
    def remove_one(self):
        return self.all_cards.pop(0)
    
    def add_cards(self,new_cards):
        #if this is a card of list then add mutiple cards by using extend
        if type(new_cards) == type([]):
            #add multiple cards
            self.all_cards.extend(new_cards)
        else:
            #add single card
            self.all_cards.append(new_cards)
    
    def __str__(self):
        return f'Player {self.name} has {len(self.all_cards)} cards.'
```

```python
# Game Setup Have to run this before starting a new game
# create players and reference player class
player_one = Player("One")
player_two = Player("Two")

#create a a full deck and shuffle
new_deck = Deck()
new_deck.shuffle()

#split the deck into two (52cards in a deck)
for x in range(26):
    player_one.add_cards(new_deck.deal_one())
    player_two.add_cards(new_deck.deal_one())
```

```python
game_on = True
round_num = 0

# while game_on
while game_on:
    round_num +=1
    print(f'Round {round_num}')
    
    if len(player_one.all_cards) == 0:
        print('Player One, out of cards. Player Two Wins!')
        game_on = False
        break
    
    elif len(player_two.all_cards) == 0:
        print('Player Two, out of cards. Player One Wins!')
        game_on = False
        break
    
    #Start a new round

    #cards that player one has on the table
    player_one_cards = []
    #for the code line below look at player class method remove_one
    player_one_cards.append(player_one.remove_one())

    player_two_cards = []
    player_two_cards.append(player_two.remove_one())

    # while at war
    
    at_war = True
    
    while at_war:
        
        if player_one_cards[-1].value > player_two_cards[-1].value:
            player_one.add_cards(player_one_cards)
            player_one.add_cards(player_two_cards)
            at_war = False
            
        elif player_one_cards[-1].value < player_two_cards[-1].value:
            player_two.add_cards(player_one_cards)
            player_two.add_cards(player_two_cards)
            at_war = False
            
        else:
            #variable for how many cards to draw out for war
            x=9
            print("War")
            if len(player_one.all_cards) < x:
                print("Player One unable to declare war")
                print("Player Two wins!")
                game_on = False
                break
            
            elif len(player_two.all_cards) < x:
                print("Player Two unable to declare war")
                print("Player One wins!")
                game_on = False
                break
                
            else:
                for num in range(x):
                    player_one_cards.append(player_one.remove_one())
                    player_two_cards.append(player_two.remove_one())
            
```
